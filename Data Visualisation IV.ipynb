```python
# Group by rev_per_mile 

# Filter the data to only include Audi, Hyundai, Suzuki, and Toyota
Manufacturer = ['Audi', 'Hyundai', 'Suzuki', 'Toyota'] 
filtered_data = car_data[car_data['Manufacturer'].isin(Manufacturer)]
# Generate the box plot
import seaborn as sns

sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_data )

# Suzuki has the highest revs per mile 
```

```python
# A Histogram of MPG in the city

# Plot histograms of MPG in the city and on the highway on the same axis
plt.hist(car_data['MPG.city'], alpha=0.5, label='City MPG') # alpha = 0.5 makes the histogram semi-transparent
plt.hist(car_data['MPG.highway'], alpha=0.5, label='Highway MPG')

# Add labels and legend
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Distribution of MPG in the City and on the Highway')
plt.legend(loc='upper right')

# Show the plot
plt.show()
# Answer to the question:
# We see Highway MPG is shifted to the right, experiencing higher MPG than City MPG 
# So this means that it is generally more fuel efficient to drive in the Highway 
# More MPG values mean more fuel efficiency - The car can travel more miles per gallon.
```

```python
# Alineplot showing the relationship between the Wheelbase and Turning Circle 

plt.plot(car_data['Wheelbase'], car_data['Turn.circle'])
plt.xlabel('Wheelbase (in.)')
plt.ylabel('Turn.circle(ft.)')
plt.title('Relationship between Wheelbase and Turning Circle')
plt.show()

# Answer to the question:
# The is a Positive relationship between wheelbase and Turning Circle 
# The relatioship is identified by the slope upwards from left to right
# So when the wheelbase gets larger , the Turning circle tends to get larger as well
```

```python
# A bar plot showing the mean horsepower for each car Type (Small, Midsize, etc.)

horsepower_by_type = car_data.groupby('Type')['Horsepower'].mean()
plt.bar(horsepower_by_type.index, horsepower_by_type.values)
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')
plt.show()

# Answer to the question:
# yes, on this case looking at the graph a larger car has more horsepower than a small car
```

```python
# A pie chart showing the distribution of engine sizes

engine_counts = car_data['EngineSize'].value_counts() # value_counts() method to count the number of occurrences of each unique engen size 
plt.pie(engine_counts.values, labels=engine_counts.index, autopct='%1.1f%')
plt.axis('equal') # sets the aspect ratio of the pie chart to be equal i.e be circular 
plt.title('Distribution of Engine Sizes')
plt.show()

# Answer to the question:
#  3.0 lITRES
```
